```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as soup
```

```python
hub_url = 'https://en.wikipedia.org/wiki/List_of_hub_airports'
hub_html = requests.get(hub_url).text
hub_soup = soup(hub_html)
```

```python
tables = hub_soup.select('table')
len(tables)
```
